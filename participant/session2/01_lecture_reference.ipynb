## What is context engineering?

**Context engineering** is the discipline of assembling the right information, in the right form,
at the right time — to ground an AI system's response.

It's more than RAG. A context engineering mindset asks:
- *What does the model need to know to answer this question correctly?*
- *How should that knowledge be retrieved, ranked, and assembled?*
- *What are the failure modes if context is missing or wrong?*

**RAG vs Context Engineering:**

| RAG | Context Engineering |
|---|---|
| Vector search → top-k chunks → prompt | Multi-source retrieval: semantic + graph + structured |
| Fixed retrieval strategy | Adaptive retrieval based on query type |
| Text chunks | Entities, relationships, metadata, structured facts |
| Good for: "What does document X say about Y?" | Good for: "How are entities X, Y, Z related, and what is the trend?" |

## The three context layers

For a financial intelligence platform, three complementary layers provide complete context:

### 1. Vector layer (semantic)
- **What it stores:** Dense embeddings of text chunks
- **What it's good at:** Similarity — "Find documents similar to this query"
- **Weakness:** No structured relationships, no explicit reasoning
- **Databricks:** Agent Bricks Knowledge Assistant, Vector Search endpoints

### 2. Graph layer (relational)
- **What it stores:** Entities and typed relationships (triplets: subject → predicate → object)
- **What it's good at:** Traversal — "Find all partners of NVIDIA's partners who mentioned supply chain risks"
- **Weakness:** Requires entity extraction upfront; sparse for novel queries
- **Databricks:** Delta tables as graph storage, Genie for natural language graph queries

### 3. Metadata layer (structured)
- **What it stores:** Typed fields — company, period, document type, revenue, category, sentiment
- **What it's good at:** Filtering and aggregation — precise, cheap, deterministic
- **Weakness:** Only what was explicitly extracted; no semantic reasoning
- **Databricks:** Delta tables, Unity Catalog, SQL warehouses

## Where each layer wins

| Query | Best layer |
|---|---|
| "What did NVIDIA say about AI infrastructure in Q3 2024?" | Vector (semantic match) |
| "Who are NVIDIA's cloud partners?" | Graph (entity traversal) |
| "Which companies have positive sentiment in FY2024 10-Ks?" | Metadata (structured filter) |
| "What are the common AI investment themes, and how do companies relate to each other?" | **All three** |

The most powerful AI assistants use all three layers together — this is the hybrid context layer.
