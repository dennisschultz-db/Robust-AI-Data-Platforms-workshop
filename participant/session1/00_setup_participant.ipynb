```python
# Create a personal schema and volume in the workshop catalog.

spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog}.{schema}")
print(f"✓ Schema ready:  {catalog}.{schema}")

spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.{schema}.financial_docs")
print(f"✓ Volume ready:  {volume_path}")
print()
print("✓ Your personal workspace is ready.")
```

## Verifying access to shared financial documents

```python
files = dbutils.fs.ls(shared_volume)
companies = [f.name.rstrip('/') for f in files if f.isDir()]

print(f"✓ Shared volume is accessible: {shared_volume}")
print(f"  Companies found: {', '.join(companies)}")

total_pdfs = 0
for company_dir in files:
    if company_dir.isDir():
        pdfs = [f for f in dbutils.fs.ls(company_dir.path) if f.name.endswith('.pdf')]
        total_pdfs += len(pdfs)
        print(f"  {company_dir.name.rstrip('/'):<12} {len(pdfs):>3} PDF(s)")

print(f"\n  Total documents: {total_pdfs}")
```

```python
# Verify the pre-built pipeline outputs are available
try:
    silver_count = spark.sql(
        f"SELECT COUNT(*) AS cnt FROM {catalog}.{shared_schema}.financial_docs_silver"
    ).collect()[0]['cnt']
    gold_count = spark.sql(
        f"SELECT COUNT(*) AS cnt FROM {catalog}.{shared_schema}.financial_docs_gold"
    ).collect()[0]['cnt']
    print(f"✓ financial_docs_silver: {silver_count:,} rows")
    print(f"✓ financial_docs_gold:   {gold_count:,} rows")
except Exception as e:
    print(f"⚠️  Pipeline outputs not yet available — instructor may need to run the pipeline first.")
    print(f"   Details: {e}")
```
